**1**.   **Listas**



```python
# Criando uma lista chamada animais

animais = [1,2,3]
animais
```

```python
animais = ['cachorro', 'gato', 12345, 6.5]
animais
```

```python
# Imprimindo o primeiro elemento da lista
animais[0]
```

```python
# Imprimindo o quarto elemento da lista
animais[3]
```

```python
# Substitindo o primeiro elemento da lista
animais[0] = 'papagaio'
animais
```

```python
# Removendo gato da lista
animais.remove('gato')
animais
```

```python
len(animais)
```

```python
'gato' in animais
```

```python
lista = [500, 30, 300, 80, 10]
```

```python
max(lista), min(lista)
```

```python
animais.append("leão")
animais
```

```python
animais.extend(['cobra', 6])
animais
```

```python
animais.count('cachorro')
```

```python
lista.sort()
lista
```


**2.   Tuplas**

```python
# As tuplas usam parentêses como sintaxe

tp = ("Banana", "Maça", 10, 50)
```

```python
# Retornando o primeiro elemento

tp[0]
```

```python
# Diferente das listas as tuplas são imutáveis, o que quer dizer que não podemos alterar seus elementos

tp[0] = 'Laranja'
```

```python
tp.count('Maça')
```

```python
tp[0:2]
```



**3.   Dicionários**



```python
# Para criar um dicionário utilizamos as chaves {}

dc = {'Maça':20, 'Banana':10, 'Laranja':15, 'Uva':5} # dicionários trabalham com o conceito de chave/valor
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc['Maça']
```

```python
# Atualizando o valor da Maça

dc['Maça'] = 25
dc
```

```python
# Retornando todas as chaves do dicionário

dc.keys()
```

```python
# Retornando todos os valores do dicionário

dc.values()
```
