```python
# Criando nosso DataFrame

df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as 5 primeiras linhas

df.head()
```

```python
# Quantidade de linhas e colunas

df.shape
```

```python
# Verificando os tipos de dados

df.dtypes
```

```python
# Qual a Receita Total?

df["Valor Venda"].sum()
```

```python
# Qual o custo total?

df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) # criando a coluna custo
```

```python
df.head()
```

```python
# Qual o custo total?

df["Custo"].sum()
```

```python
# Agora que temos a receita e o custo total, podemos achar o lucro total. Vamos criar uma coluna de Lucro que será Receita - Custo.

df["Lucro"] = df["Valor Venda"] -df["Custo"]
```

```python
df.head()
```

```python
# Total Lucro

round(df["Lucro"].sum(),2)
```

```python
# Criando uma coluna com total de dias para enviar o produto

df["Tempo_Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(5)
```

```python
df["Tempo_Envio"].dtype
```

```python
# Média de tempo de envio por Marca

df.groupby(df["Marca"])["Tempo_Envio"].mean()
```

**Missing Values**

```python
# Verificando se temos dados faltantes

df.isnull().sum()
```

**E se quisermos saber o Lucro por Ano e Marca?**

```python
# Vamos agrupar por Ano e Marca

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Resetando o index

lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos?

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico total de produtos vendidos

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
# Selecionando apenas as vendas de 2019

df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Marca x Lucro")
plt.xlabel("Marca")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Classe x Lucro")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```
